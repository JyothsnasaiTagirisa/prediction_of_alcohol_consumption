# replacing the objects into numarics

```python
df['STATEININDIA'].replace(['Kerala','Tamil Nadu','karnataka','Andhra Pradesh','Telangana','Goa','Puducherry'],[0,1,2,3,4,5,6],inplace=True)
```

```python
df
```

# considering feature variables and target variable

```python
X=df[['YEAR','STATEININDIA','UNEMPLOYMENT','TAXONCOUNTRYALCOHOL','REVENUEGENERATEDOFCOUNTRYALCOHOL']]
y=df[['CONSUMTIONOFCONTRYAOCOHOL']]
```

# statsmodel

```python
import statsmodels.api as sm
model = sm.OLS(y,X).fit()
predictions = model.predict(X)
model.summary()
```

# Multi linear regression with 80% for testing data and 20% for traning data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
X_train=X_train.values.reshape((-1,5))
```

```python
X_train
```

```python
from sklearn import linear_model as lm
model=lm.LinearRegression()
results=model.fit(X_train,y_train)
```

```python
results
```

```python
#score function
accuracy=model.score(X_train,y_train)
print("Accuracy of the model:",accuracy)
```

```python
#print the coff
print("intercept:",model.intercept_)
print("slope",model.coef_)
```

```python
X_test=X_test.values.reshape((-1,5))
```

```python
X_test
```

```python
#predictions for the model
predictions=model.predict(X_test)
print("predicted consumption",predictions,sep="\n")
```

```python
plt.scatter(y_test,predictions)
```

# variations in consumption can be known from predictions

```python
#make a prediction
Xnew=[[2020,4,15.6,130,690233]]
#make a prediction
ynew=model.predict(Xnew)
```

```python
ynew
```

```python
from sklearn.metrics import mean_squared_error, r2_score
X_train = X_train.reshape(-1,5)
y_train_predictions = model.predict(X_train)

X_test = X_test.reshape(-1,5)
y_test_predictions = model.predict(X_test)
```

```python
# printing values
print('Slope:' ,model.coef_)
print('Intercept:', model.intercept_)
print("\n")

# model evaluation for training set
import numpy as np
rmse_training = (np.sqrt(mean_squared_error(y_train, y_train_predictions)))
r2_training = r2_score(y_train, y_train_predictions)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse_training))
print('R2 score is {}'.format(r2_training))
print("\n")

# model evaluation for testing set
rmse_testing = (np.sqrt(mean_squared_error(y_test, y_test_predictions)))
r2_testing = r2_score(y_test, y_test_predictions)

print("The model performance for testing set")
print("--------------------------------------")
print('Root mean squared error: ', rmse_testing)
print('R2 score: ', r2_testing)
```
